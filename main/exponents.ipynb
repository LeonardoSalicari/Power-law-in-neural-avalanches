```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dt = np.array([0.0003, 0.0005, 0.001, 0.003, 0.005, 0.01])
# sizes
tau = np.array([1.908, 1.823, 1.738, 1.459, 1.294, 1.197])
# duration
alpha = np.array([2.169, 2.007, 1.948, 1.611, 1.415, 1.332])
# gamma fit
gamma_fit = np.array([1.250, 1.287, 1.288, 1.376, 1.445, 1.528])
gamma_fit_err = np.array([0.004, 0.006, 0.003, 0.006, 0.005, 0.006])
# gamma from Crackling
gamma_c = np.array([1.29, 1.22, 1.29, 1.33, 1.41, 1.68])
gamma_c_err = np.array([0.08, 0.06, 0.05, 0.05, 0.03, 0.03])
```

```python
diff_gamma = gamma_fit - gamma_c
diff_gamma_err = np.sqrt( gamma_fit_err**2 + gamma_c_err**2 )

font_size = 15

fig, ax = plt.subplots(nrows=1, ncols=1)#, figsize=(8,6))
ax.errorbar(dt, diff_gamma, yerr = diff_gamma_err, fmt='o', capsize=4)
ax.grid(axis='y', linestyle='--')
ax.axhline(c='r')
ax.set_xlabel('$dt$', fontsize = font_size)
ax.set_ylabel('$\gamma_{\,fit} - \gamma_c$', fontsize = font_size)
ax.set_xscale('log')
#ax.axes.get_xaxis().set_visible(False)
plt.savefig('img_exp/crackling_rel', format='pdf')
```

```python
fig, ax1 = plt.subplots(nrows=1, ncols=1, figsize=(10,7))
ax1.plot(dt, alpha, 'o--', label ='alpha' )
ax1.plot(dt, tau, 'o--', label ='tau' )
ax1.grid(axis='y', linestyle='--')
ax1.set_xlabel('$dt$', fontsize = font_size)
ax1.set_ylabel('Exponents', fontsize = font_size)
ax1.set_xscale('log')
ax1.set_xticks(dt)
ax1.legend()
plt.savefig('img_exp/exponent_vs_dt', format='pdf')
```
